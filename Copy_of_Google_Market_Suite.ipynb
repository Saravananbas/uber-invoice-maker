```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown


def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

**Chat Models**

The `ChatGoogleGenerativeAI` class is the primary interface for interacting with Google's Gemini chat models. It allows users to send and receive messages using a specified Gemini model, suitable for various conversational AI applications.

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-pro", google_api_key=GOOGLE_API_KEY)
tweet_prompt = PromptTemplate.from_template("You are a content creator. Write me a tweet about {topic}.")
tweet_chain = LLMChain(llm=llm, prompt=tweet_prompt, verbose=True)
```

# Builld Application using Gradio for a quick PoC and Demo

```python
import gradio as gr
from langchain_google_genai.llms import GoogleGenerativeAI

# Initialize the GoogleGenerativeAI model
llm = GoogleGenerativeAI(model="gemini-pro", google_api_key=GOOGLE_API_KEY)

def generate_output(input_text):
    # Invoke the model to generate output based on input text
    output_text = llm.invoke(input_text)
    # Convert the output to markdown format
    markdown_output = f"**Generated Output:**\n\n{output_text}"
    return markdown_output

# Define the input and output components for the Gradio app
input_textbox = gr.Textbox(lines=10, label="Enter text here")
output_textbox = gr.Markdown("")

# Create the Gradio interface
gr.Interface(fn=generate_output, inputs=input_textbox, outputs=output_textbox, title="Text to Markdown Generator").launch()
```
